```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json, os
plt.rcParams['figure.figsize'] = 12,8
```

```python
# import everything into a list directly from the raw json files 
flickr_yahoo_path = 'flickr_yahoodata'
photos = []
for fnum in range(10):
    json_file_name = "yfcc100m_dataset-%d_nyc.json" % fnum
    print json_file_name
    with open(os.path.join(flickr_yahoo_path,json_file_name)) as fp:
        photos.extend(json.load(fp))
```

```python
# convert into a data frame 
photo_df = pd.DataFrame(photos)
photo_df.head()
```

```python
# limit data by a certain radius 
init_loc = (40.74844,-73.985664) # empire state building, latitude/longitude
radius = 0.01
idx = sqrt((photo_df['lng']-init_loc[1])**2 + (photo_df['lat']-init_loc[0])**2) < radius
# plot data 
plt.plot(photo_df['lng'][idx], photo_df['lat'][idx],'.k')
```

```python
# cluster to find centroids 
#from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
data = photo_df[['lng','lat']][idx].values
#estimator = KMeans(init='k-means++', n_clusters=10, n_init=10)
#estimator.fit(data)
#centroids = estimator.cluster_centers_
#plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=169, linewidths=2, color='r', zorder=10)
estimator = DBSCAN(eps=0.0005, min_samples=1000)
estimator.fit(data)

core_samples_mask = np.zeros_like(estimator.labels_, dtype=bool)
core_samples_mask[estimator.core_sample_indices_] = True
labels = estimator.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

# Black removed and is used for noise instead.
unique_labels = set(labels)
centroids = []
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = data[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=8)
    if k != -1:
        centroid = xy.mean(axis=0)
        centroids.append(centroid)
        plt.plot(centroid[0],centroid[1],'o',markersize=16)

    xy = data[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

```python
# round to the nearest hour or whatever other interval 
# http://stackoverflow.com/questions/24479577/pandas-timestamp-index-rounding-to-the-nearest-5th-minute
nsround = 30*60*1000000000   # 60 minutes in nanoseconds
photos2.index = pd.DatetimeIndex(((photos2.index.astype(np.int64) // nsround + 1 ) * nsround))
# do the same thing for the date_taken column
photos2['date_taken'] = pd.DatetimeIndex(((pd.to_datetime(photos2.date_taken).astype(np.int64) // nsround + 1 ) * nsround))
photos2 = photos2.drop_duplicates(['user_id','date_taken'])
print photos2.shape[0]
photos2.head()
```

```python
# get the hour mean
# http://stackoverflow.com/questions/16967165/getting-the-average-of-a-certain-hour-on-weekdays-over-several-years-in-a-pandas 
def get_hour_mean(photo_df, loc, radius=0.005, normalize=False):
    idx = np.sqrt((photo_df['lat']-loc[0])**2 + (photo_df['lng']-loc[1])**2) < radius
    print 'number of photos near location ', sum(idx)
    hour_mean = photo_df.user_id[idx].groupby(lambda x: x.hour).count()
    if normalize:
        hour_mean = hour_mean / float(photo_df.user_id[idx].count())
    return hour_mean

# compute the hour mean for all of nyc
hour_all_normalized = photos2.user_id.groupby(lambda x: x.hour).count()/photos2.user_id.count()
hour_all = photos2.user_id.groupby(lambda x: x.hour).count()

time_axis = range(24)
plt.plot(time_axis, hour_all)
```

```python
# getting the hour mean of favorite counts at a location 
import pymysql as mdb
db = mdb.connect('localhost', 'root', '', 'insight')
def get_favcounts_sql(db,loc,radius=0.005):
    with db:
        cur = db.cursor(mdb.cursors.DictCursor)
        cmd = "SELECT flickr_favorites.Id,Fav,Lat,Lng, date_taken FROM flickr_yahoo_nyc JOIN flickr_favorites \
        ON flickr_yahoo_nyc.Id = flickr_favorites.Id\
        WHERE ( POWER(lat - (%s), 2) + POWER(lng - (%s), 2) ) < POWER(%s, 2)" % \
        (loc[0], loc[1], radius)
        #print cmd
        cur.execute(cmd)
        results = cur.fetchall()
    print len(results)
    results = pd.DataFrame(results)
    # round to the nearest half hour
    results.index = pd.DatetimeIndex(((results.date_taken.astype(np.int64) // nsround + 1 ) * nsround))
    # get the average favorite counts
    #results = results.Fav[results.Fav>0].groupby(lambda x: (x.hour, x.minute)).count()
    return results

# test case 
init_loc = [40.74844,-73.985664]
results = get_favcounts_sql(db,init_loc,radius=0.01)
results.head()
```

```python
# test smoothing function 
from scipy.ndimage import convolve

# regular convolution in numpy -- doesn't have circular
def movingaverage(interval, window_size):
    window= np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')

# cconv can be gotten from scipy
def movingaverage2(interval, window_size):
    window= np.ones(int(window_size))/float(window_size)
    return convolve(interval, window, mode='wrap')

hour_loc = get_hour_mean(photos2,[40.74844,-73.985664],normalize=normalize)/hour_all
hour_loc_av = movingaverage(hour_loc, 3)
hour_loc_av2 = movingaverage2(hour_loc, 3)
fig = plt.figure(figsize=(6,3))
plt.plot(time_axis,hour_loc,'k')
#plt.plot(time_axis,hour_loc_av,'r')
plt.plot(time_axis,hour_loc_av2,'b')
```

```python
# test gaussian weighting function 
def gauss2(X, r0=(0,0), sigma=1):
    rad_sq = np.sum(np.square(np.array(X)-np.array(r0)))
    return math.exp(-(rad_sq)/(2*sigma**2))

def gauss(X, r0=0, sigma=1):
    return math.exp(-(X-r0)**2/(2*sigma**2))

fig = plt.figure(figsize=(6,3))
x = linspace(-3,3,50)
plt.plot(x, [gauss(y) for y in x])
plt.plot(x, [gauss2((0,y)) for y in x],'r')
```

```python
# write a function for getting the favorite counts at a location by doing a join
def get_favcounts_sql(db,init_loc,lim=0.005):
    with db:
        cur = db.cursor(mdb.cursors.DictCursor)
        cmd = "SELECT flickr_favorites.Id,Fav,Lat,Lng FROM flickr_yahoo_nyc JOIN flickr_favorites \
        ON flickr_yahoo_nyc.Id = flickr_favorites.Id\
        WHERE ((lat BETWEEN %s AND %s) AND (lng BETWEEN %s AND %s))" % \
        (init_loc[0]-lim,init_loc[0]+lim,init_loc[1]-lim,init_loc[1]+lim)
        cur.execute(cmd)
        results = cur.fetchall()
    return results

fav_counts = pd.DataFrame(get_favcounts_sql(db, [40.74844,-73.985664]))
fav_counts.head()
fav_counts['Fav'].describe()
```

```python
# different way to calculate favorite counts 
# use the reduced data set with duplicate users taken out for each half hour interval
# user the cluster labels to restrict photos 
def get_favcounts_sql2(db,clusterid):
    with db:
        cur = db.cursor(mdb.cursors.DictCursor)
        cmd = "SELECT flickr_favorites.Id, Fav FROM flickr_yahoo_nyc2 JOIN flickr_favorites \
        ON flickr_yahoo_nyc2.Id = flickr_favorites.Id\
        WHERE cluster_label = %d" % clusterid
        cur.execute(cmd)
        results = cur.fetchall()
    return results

fav_counts = get_favcounts_sql2(db,0)
fav_counts = pd.DataFrame(fav_counts)
fav_counts.head()
#print fav_counts['Fav'].describe()
```

```python
%reload_ext autoreload
from flickr_sites import *
from tripomatic_lookup import *
from scipy.stats import pearsonr
# test correlation between photo density and touristiness score 

db = mdb.connect('localhost', 'root', '', 'insight')
#curr_loc = [40.74844,-73.985664]
# get all possible attractions 
#attractions = get_tripomatic_sql(db,curr_loc[0],curr_loc[1],lim=0.5)
#print len(attractions), 'attractions'

# load all centroids 
with db:
    cur = db.cursor(mdb.cursors.DictCursor)
    cmd = "SELECT * FROM flickr_clusters_nyc2"
    cur.execute(cmd)
    centroids = cur.fetchall()

all_score = []
all_density = []
all_fav = []
for cent in centroids:
    curr_loc = (cent['lat'],cent['lng'])
    # attractions within neighborhood
    attractions = get_tripomatic_sql(db,curr_loc[0],curr_loc[1],lim=0.01)
    # touristiness score
    tripomatic_score = touristy_score(curr_loc, attractions)
    
    # photo density 
    #photos = get_heatmap_sql(db,curr_loc,lim=0.005)
    #photo_density = len(photos)
    photo_density = cent['nphotos']
    
    # average favorite count for this location
    #favs = get_favcounts_sql(db, curr_loc)
    favs = get_favcounts_sql2(db, cent['index'])
    print cent['index'], len(favs)
    if favs:
        fav_counts = pd.DataFrame(favs)
        all_fav.append(fav_counts['Fav'].mean())
    else:
        all_fav.append(0.)

    #print photo_density, tripomatic_score
    all_density.append(photo_density)
    all_score.append(tripomatic_score)

all_score = np.array(all_score)
all_density = np.array(all_density)
all_fav = np.array(all_fav)
```
