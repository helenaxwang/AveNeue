```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json, os
plt.rcParams['figure.figsize'] = 12,8
```

```python
# import everything into a list directly from the raw json files 
flickr_yahoo_path = 'flickr_yahoodata'
photos = []
for fnum in range(10):
    json_file_name = "yfcc100m_dataset-%d_nyc.json" % fnum
    print json_file_name
    with open(os.path.join(flickr_yahoo_path,json_file_name)) as fp:
        photos.extend(json.load(fp))
```

```python
# convert into a data frame 
photo_df = pd.DataFrame(photos)
photo_df.head()
```

```python
# limit data by a certain radius 
init_loc = (40.74844,-73.985664) # empire state building, latitude/longitude
radius = 0.01
idx = sqrt((photo_df['lng']-init_loc[1])**2 + (photo_df['lat']-init_loc[0])**2) < radius
# plot data 
plt.plot(photo_df['lng'][idx], photo_df['lat'][idx],'.k')
```

```python
# cluster to find centroids 
#from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
data = photo_df[['lng','lat']][idx].values
#estimator = KMeans(init='k-means++', n_clusters=10, n_init=10)
#estimator.fit(data)
#centroids = estimator.cluster_centers_
#plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=169, linewidths=2, color='r', zorder=10)
estimator = DBSCAN(eps=0.0005, min_samples=1000)
estimator.fit(data)

core_samples_mask = np.zeros_like(estimator.labels_, dtype=bool)
core_samples_mask[estimator.core_sample_indices_] = True
labels = estimator.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)

# Black removed and is used for noise instead.
unique_labels = set(labels)
centroids = []
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = data[class_member_mask & core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=8)
    if k != -1:
        centroid = xy.mean(axis=0)
        centroids.append(centroid)
        plt.plot(centroid[0],centroid[1],'o',markersize=16)

    xy = data[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
plt.show()
```

```python
# round to the nearest hour
# http://stackoverflow.com/questions/24479577/pandas-timestamp-index-rounding-to-the-nearest-5th-minute
nsround = 30*60*1000000000   # 60 minutes in nanoseconds
photos2.index = pd.DatetimeIndex(((photos2.index.astype(np.int64) // nsround + 1 ) * nsround))
# do the same thing for the date_taken column
photos2['date_taken'] = pd.DatetimeIndex(((pd.to_datetime(photos2.date_taken).astype(np.int64) // nsround + 1 ) * nsround))
photos2 = photos2.drop_duplicates(['user_id','date_taken'])
print len(photos2)
photos2.head(n=8)
```

```python
# get the hour mean
# http://stackoverflow.com/questions/16967165/getting-the-average-of-a-certain-hour-on-weekdays-over-several-years-in-a-pandas 
def get_hour_mean(photo_df, loc, radius=0.005, normalize=False):
    idx = np.sqrt((photo_df['lat']-loc[0])**2 + (photo_df['lng']-loc[1])**2) < radius
    print 'number of photos near location ', sum(idx)
    hour_mean = photo_df.user_id[idx].groupby(lambda x: x.hour).count()
    if normalize:
        hour_mean = hour_mean / float(photo_df.user_id[idx].count())
    return hour_mean

# compute the hour mean for all of nyc
hour_all_normalized = photos2.user_id.groupby(lambda x: x.hour).count()/photos2.user_id.count()
hour_all = photos2.user_id.groupby(lambda x: x.hour).count()

time_axis = range(24)
plt.plot(time_axis, hour_all)
```

```python
from scipy.ndimage import convolve

# regular convolution in numpy -- doesn't have circular
def movingaverage(interval, window_size):
    window= np.ones(int(window_size))/float(window_size)
    return np.convolve(interval, window, 'same')

# cconv can be gotten from scipy
def movingaverage2(interval, window_size):
    window= np.ones(int(window_size))/float(window_size)
    return convolve(interval, window, mode='wrap')

hour_loc = get_hour_mean(photos2,allLocs[0],normalize=normalize)/hour_all
hour_loc_av = movingaverage(hour_loc, 3)
hour_loc_av2 = movingaverage2(hour_loc, 3)
plt.plot(time_axis,hour_loc,'k')
#plt.plot(time_axis,hour_loc_av,'r')
plt.plot(time_axis,hour_loc_av2,'b')
```
